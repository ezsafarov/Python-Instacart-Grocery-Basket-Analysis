```python
# Assign 'Regular customer' to customers with 11 to 40 orders
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Assign 'New customer' to customers with 10 or fewer orders
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Calculate the mean price of products purchased by each user
user_avg_spending = ords_prods_merge.groupby('user_id')['prices'].mean()

# Map the user_avg_spending to the original DataFrame
ords_prods_merge['avg_spending'] = ords_prods_merge['user_id'].map(user_avg_spending)

# Create a spending flag based on the avg_spending column
ords_prods_merge['spending_flag'] = 'High spender'  # Default value for all users
ords_prods_merge.loc[ords_prods_merge['avg_spending'] < 10, 'spending_flag'] = 'Low spender'

# Verify the results
print(ords_prods_merge[['user_id', 'avg_spending', 'spending_flag']].head())
```

```python
# Count the number of users in each spending category
spending_counts = ords_prods_merge['spending_flag'].value_counts()

print(spending_counts)
```

```python
# Calculate the median of 'days_since_prior_order' for each user
user_order_median = ords_prods_merge.groupby('user_id')['days_since_prior_order'].median()

# Map the median order frequency back to the original DataFrame
ords_prods_merge['order_frequency'] = ords_prods_merge['user_id'].map(user_order_median)

# Create a frequency flag based on the median days_since_prior_order
ords_prods_merge['frequency_flag'] = 'Non-frequent customer'  # Default value
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'

# Verify the results
print(ords_prods_merge[['user_id', 'order_frequency', 'frequency_flag']].head())
```

```python
# Count the number of users in each frequency category
frequency_counts = ords_prods_merge['frequency_flag'].value_counts()

print(frequency_counts)
```
